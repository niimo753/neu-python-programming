## Exercise 1


```python
a = float(input("a = "))
b = float(input("b = "))
n = int(input("n = "))
while n%2==1:
    n = int(input("> n must be an EVEN number. Try again: n = "))
delx = (b - a)/n
inte = delx/3 * (1/a + 1/b)
for i in range(1, n):
    inte += (delx/3) * (1/(a+i*delx)) * (4-((i-1)%2)*2)
print("\n>>> The integral is about:", inte)
```

## Exercise 3
$$ A(x) = 1 + 2x + 3x^2 + ... + nx^{n-1}$$

```python
x = float(input("x = "))
n = int(input("n = "))
a = 0
for i in range(1, n+1):
    a += i*(x**(i-1))
print("\n>>> With n = {}, A({}) = {}".format(n, x, a))
```
